```python
# wav 파일 경로 list 에 저장
import glob
import os

# 특정 폴더 경로
folder_path = './segmented_character_voice/masiro'

# 해당 폴더에 있는 모든 .wav 파일 경로를 리스트로 저장
wav_files = glob.glob(os.path.join(folder_path, '*.wav'))

# 파일 이름만 추출 및 상대 경로 맞추기
wav_file_names = ["./segmented_character_voice/masiro/" + os.path.basename(file) for file in wav_files]

# 결과 출력
wav_file_names.sort()
print(wav_file_names[:10])
```

```python
import ipywidgets as widgets
from IPython.display import display, Audio
import librosa
import matplotlib.pyplot as plt
import numpy as np
import os

# WAV 파일 및 텍스트 데이터 준비 (예시로 10개의 파일)
wav_files = wav_file_names  # 실제 경로의 wav 파일 리스트
txt_files = lines  # 실제 텍스트 데이터 리스트 (라인 단위로 저장)

# 텍스트 파일 경로 (실제 텍스트 파일의 경로)
text_file_path = 'your_text_file.txt'  # 해당 파일의 경로로 바꿔주세요

# 현재 파일 인덱스 저장
current_index = 0

# WAV 플레이어와 텍스트 출력을 위한 Output 위젯 생성
output_audio = widgets.Output()
output_waveform = widgets.Output()
warning_output = widgets.Output()  # 경고 메시지를 출력할 Output

# 현재 페이지/전체 페이지를 표시할 Label 생성
page_label = widgets.Label()

# WAV 플레이어와 텍스트 출력 초기화
text_output = widgets.HTML(f"<h2>대사: </h2><p></p>")

# WAV 플레이어와 텍스트 출력 UI 업데이트 함수
def update_display(index):
    global output_audio, output_waveform, warning_output, text_output, wav_files, txt_files, page_label
    if len(wav_files) > 0:  # 파일이 있을 때만 실행
        text_split = txt_files[index].split(" ", 1)
        wav_filename_from_text = text_split[0].split(" ", 1)[0].split("|")[0]

        # 조건 체크: text_split[0]와 wav_files[index]가 일치하는지 확인
        if wav_filename_from_text == wav_files[index]:
            # 텍스트 업데이트
            text_output.value = f"<h2>대사: {text_split[1]}</h2><p>{text_split[0]}</p>"

            # 오디오 출력 갱신
            with output_audio:
                output_audio.clear_output()  # 기존 출력 내용 삭제
                display(Audio(wav_files[index], autoplay=False))  # 새로운 오디오 파일 재생

            # 오디오 파형 시각화
            with output_waveform:
                output_waveform.clear_output()  # 기존 파형 지우기
                plot_waveform(wav_files[index])  # 새로운 파형 그리기

            # 경고 메시지 초기화
            with warning_output:
                warning_output.clear_output()  # 경고 메시지 지우기

        else:
            # 기존 출력 (텍스트, 오디오, 파형) 지우기
            text_output.value = "<h2>대사: </h2><p></p>"
            with output_audio:
                output_audio.clear_output()  # 오디오 지우기
            with output_waveform:
                output_waveform.clear_output()  # 파형 지우기

            # 경고 메시지 출력
            with warning_output:
                warning_output.clear_output()  # 기존 경고 삭제
                print(f"경고: 파일 이름이 일치하지 않습니다.")
                print(f"텍스트에서 추출한 이름: {wav_filename_from_text}")
                print(f"WAV 파일 이름: {wav_files[index]}")

        # 페이지 상태 갱신 ('페이지/전체 페이지' 형식)
        page_label.value = f"{index + 1}/{len(wav_files)}"
    else:
        # 파일이 더 이상 없으면 UI 초기화
        text_output.value = "<h2>더 이상 항목이 없습니다</h2>"
        output_audio.clear_output()
        output_waveform.clear_output()
        page_label.value = "0/0"

# 오디오 파일의 파형을 그리는 함수
def plot_waveform(wav_file):
    y, sr = librosa.load(wav_file, sr=None)  # 오디오 데이터 로드
    plt.figure(figsize=(10, 3))
    plt.plot(np.linspace(0, len(y)/sr, len(y)), y)  # 직접 파형 그리기
    plt.title(f"Waveform of {wav_file}")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.show()

# '다음' 버튼 클릭 핸들러 함수
def on_next_clicked(b):
    global current_index
    if current_index < len(wav_files) - 1:
        current_index += 1
        update_display(current_index)

# '이전' 버튼 클릭 핸들러 함수
def on_prev_clicked(b):
    global current_index
    if current_index > 0:
        current_index -= 1
        update_display(current_index)

# '삭제' 버튼 클릭 핸들러 함수
def on_delete_clicked(b):
    global current_index, wav_files, txt_files
    if wav_files and txt_files:  # 리스트에 항목이 있는 경우
        
        # 실제로 .wav 파일 삭제
        wav_file_to_delete = wav_files[current_index]
        if os.path.exists(wav_file_to_delete):
            os.remove(wav_file_to_delete)  # 실제 파일 삭제
            print(f"Deleted WAV file: {wav_file_to_delete}")
        
        # 텍스트 파일에서 해당 줄 삭제
        txt_line_to_delete = txt_files[current_index]
        delete_line_from_file(text_file_path, txt_line_to_delete)
        print(f"Deleted line from text file: {txt_line_to_delete}")
        
        # 리스트에서 항목 삭제
        del wav_files[current_index]  # 현재 인덱스의 wav 파일 삭제
        del txt_files[current_index]  # 현재 인덱스의 텍스트 파일 삭제
        
        if current_index >= len(wav_files):  # 인덱스가 리스트 범위를 초과하지 않도록 조정
            current_index = max(0, len(wav_files) - 1)
        
        update_display(current_index)

# 버튼 생성
prev_button = widgets.Button(description="이전")
next_button = widgets.Button(description="다음")
delete_button = widgets.Button(description="삭제")

# 버튼 클릭 시 이벤트 처리 연결
prev_button.on_click(on_prev_clicked)
next_button.on_click(on_next_clicked)
delete_button.on_click(on_delete_clicked)

# 버튼과 UI 배치
button_box = widgets.HBox([prev_button, next_button, delete_button])

# 텍스트와 WAV, 파형, 경고 메시지를 보여주는 박스
display_box = widgets.VBox([text_output, output_audio, output_waveform, warning_output])

# 전체 UI 레이아웃 설정
hbox = widgets.VBox([display_box, button_box, page_label], layout=widgets.Layout(width='100%'))

# UI 디스플레이
display(hbox)

# 첫 번째 WAV 파일과 텍스트 표시
update_display(current_index)
```

```python
import os
import numpy as np
import torch
from torch import no_grad, LongTensor
import argparse
import commons  # 일반적인 유틸리티 모듈
from mel_processing import spectrogram_torch  # 멜 스펙트로그램 생성 함수
import utils  # 다양한 유틸리티 함수 모음
from models import SynthesizerTrn  # 텍스트를 음성으로 변환하는 모델
import gradio as gr  # 웹 기반 인터페이스를 위한 Gradio 모듈
import librosa  # 오디오 처리를 위한 라이브러리
import webbrowser  # 웹브라우저를 열기 위한 모듈

from text import text_to_sequence, _clean_text  # 텍스트를 처리하고 시퀀스로 변환하는 함수
device = "cuda:0" if torch.cuda.is_available() else "cpu"  # GPU가 있으면 CUDA 장치를 사용하고, 없으면 CPU 사용

# 로그 레벨 설정 (경고만 출력)
import logging
logging.getLogger("PIL").setLevel(logging.WARNING)
logging.getLogger("urllib3").setLevel(logging.WARNING)
logging.getLogger("markdown_it").setLevel(logging.WARNING)
logging.getLogger("httpx").setLevel(logging.WARNING)
logging.getLogger("asyncio").setLevel(logging.WARNING)

# 언어 마크 설정
language_marks = {
    "Japanese": "",  # 일본어는 빈 문자열로 처리
    "日本語": "[JA]",  # 일본어
    "简体中文": "[ZH]",  # 중국어
    "English": "[EN]",  # 영어
    "한국어": "[KO]",  # 한국어
    "Mix": "",  # 혼합
}

# 언어 목록
lang = ['日本語', '简体中文', 'English', '한국어', 'Mix']
```

```python
# 텍스트를 모델 입력에 맞게 변환하는 함수
def get_text(text, hps, is_symbol):
    # 텍스트를 시퀀스로 변환
    text_norm = text_to_sequence(text, hps.symbols, [] if is_symbol else hps.data.text_cleaners)
    if hps.data.add_blank:
        text_norm = commons.intersperse(text_norm, 0) # 블랭크 추가 (음소 간)
    text_norm = LongTensor(text_norm) # 텐서로 변환
    return text_norm

# 텍스트를 음성으로 변환하는 함수 생성
def create_tts_fn(model, hps, speaker_ids):
    def tts_fn(text, speaker, language, speed): # 텍스트, 캐릭터, 언어, 속도
        if language is not None:
            # 언어 마크를 텍스트 앞뒤에 추가
            text = language_marks[language] + text + language_marks[language]
        speaker_id = speaker_ids[speaker] # 선택된 스피커 ID
        stn_tst = get_text(text, hps, False) # 텍스트를 모델 입력에 맞게 변환
        with no_grad(): # 그래디언트 계산 비활성화 (평가 모드)
            x_tst = stn_tst.unsqueeze(0).to(device) # 배치 차원 추가 후 GPU로 전송
            x_tst_lengths = LongTensor([stn_tst.size(0)]).to(device) # 텍스트 길이 정보
            sid = LongTensor([speaker_id]).to(device) # 스피커 ID 정보
            # 모델을 통해 음성 생성 (속도 조절 가능)
            audio = model.infer(x_tst, x_tst_lengths, sid=sid, noise_scale=.667, noise_scale_w=0.8,
                                length_scale=1.0 / speed)[0][0, 0].data.cpu().float().numpy()
        del stn_tst, x_tst, x_tst_lengths, sid # 메모리 해제
        return "Success", (hps.data.sampling_rate, audio) # 성공 메시지 및 생성된 오디오 반환

    return tts_fn
```

```python
# 명령행 인자 파싱
# parser.add_argument("--share", default=False, help="make link public (used in colab)")

# args = parser.parse_args()
hps = utils.get_hparams_from_file("./finetune_speaker.json")

# 모델 로드 및 설정
net_g = SynthesizerTrn(
    len(hps.symbols),
    hps.data.filter_length // 2 + 1,
    hps.train.segment_size // hps.data.hop_length,
    n_speakers=hps.data.n_speakers,
    **hps.model).to(device)
_ = net_g.eval()  # 평가 모드로 전환

# 모델 체크포인트 로드
_ = utils.load_checkpoint("./OUTPUT_MODEL/G_latest.pth", net_g, None)
speaker_ids = hps.speakers  # 스피커 ID 로드
speakers = list(hps.speakers.keys())  # 스피커 목록 생성
```

```python
# TTS 함수와 음성 변환 함수 생성
tts_fn = create_tts_fn(net_g, hps, speaker_ids)
```
